```python
#importing the modules we will use in this study
import numpy as np
import pandas as pd 
import sqlite3
import matplotlib.pyplot as plt
```

```python
# initializing the database
database = 'covid.db'
```

```python
# connecting the database
conn = sqlite3.connect(database)
```

```python
# display of the countries with the highest number of vaccinations that administered only Pfizer/BioNTech vaccine
pfizer=pd.read_sql("""SELECT country_name, data_vaccination, iso_code, vaccines_name,total_vaccinations
                        FROM country_vaccinations
                        WHERE vaccines_name='Pfizer/BioNTech'
                        ORDER BY total_vaccinations DESC;""", conn)
```

```python
display(pfizer.head(10).style.background_gradient(cmap="YlGn"))
plt.show()
```
